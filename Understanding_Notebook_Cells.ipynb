# Understanding Notebook Cells: Code vs Markdown

This notebook demonstrates the difference between Code cells and Markdown cells, and shows how to use them intentionally for readability and collaboration.

## Why use Markdown cells?

Markdown cells are used to explain what the notebook does, describe the purpose of code, and provide structure using headings, paragraphs, and lists.

```python
# This is a simple code cell that performs a calculation
a = 5
b = 7
a + b
```

## What this code cell does

The code cell above defines two variables and computes their sum. The output is produced when the code cell is executed.

```python
# Another code cell showing a Python list and its length
items = ['apple', 'banana', 'cherry']
len(items)
```

## Why this matters

Use code cells for execution and Markdown cells for explanation. This separation makes notebooks easier to read, review, and maintain.

## Notebook structure summary

- Title and purpose in a Markdown cell
- Explanations and context in Markdown cells
- Python commands in Code cells
- Code outputs shown only after executing Code cells

## Kernel Control and Execution Order

This notebook now includes a kernel control section. Use this part to practice:

- Executing cells in order
- Interrupting a long-running cell
- Restarting the kernel and rerunning from the top

Restarting clears all variables and hidden state, so the notebook becomes reproducible only when cells are run from the start.

```python
# Cell 1: Define a variable and show its value
x = 42
x
```

```python
# Cell 2: Use the variable from the previous cell
# This cell depends on Cell 1 having run first.
x * 3
```

### Why execution order matters

The cell above depends on the value of `x` defined earlier. If the kernel is restarted and this cell is executed without rerunning the cell that defines `x`, it will fail with a `NameError`.

This is the hidden state problem: the notebook engine remembers variables only while the kernel is running.

```python
# Cell 3: Long-running operation to practice interrupting
import time
for i in range(10):
    print(f"Step {i+1}/10")
    time.sleep(1)

print("This cell finished normally.")
```

### Interrupting execution

Run the cell above and then use the notebook's interrupt action if it runs longer than expected. The kernel should stop the loop and keep the notebook responsive.

If the notebook becomes unresponsive, use restart instead.

```python
# Cell 4: Confirm notebook is still responsive after interrupt or restart
print("Notebook is responsive.")
```

### Restarting the kernel

After restarting the kernel, the variable `x` is cleared from memory. To restore the notebook state, run all cells from the top.

This ensures that someone else opening the notebook can reproduce the results without hidden execution order dependencies.
